```python
#多階
import numpy as np
import pandas as pd

data = pd.Series(np.random.randn(9),
                 index =[['a','a','a','b','b','b','c','c','d'],
                         [1,2,3,1,2,3,1,2,3]])
type(data.index)
data.index.names = ('key1','key2')
data
```

```python
import numpy as np
import pandas as pd

data = pd.Series(np.random.randn(9),
                 index =[['a','a','a','b','b','b','c','c','d'],
                         [1,2,3,1,2,3,1,2,3]])
type(data.index)
data.index.names = ('key1','key2')
data.unstack()
```

```python

data.unstack(level='key2')
```

```python
data.unstack(level='key1')
```

```python
data.unstack(level='key1',fill_value=0)
```

```python
#多階層的DataFrame
import numpy as np
import pandas as pd

df = pd.DataFrame(np.arange(12).reshape((4,3)),
                  index=[['a','a','b','b'],[1, 2, 1, 2]],
                  columns=[['台北','台北','台中'],['Green','Red','Green']])
df.index.names=['key1','key2']
df.columns.names = ['縣市',"顏色"]
df
```

```python
df1 = df.stack(level=['縣市','顏色'],future_stack=True)
df1
```

```python
df1.unstack(level=['key1','縣市'],fill_value=0)
```

```python

df1.unstack(level=['key1','縣市','key2'],fill_value=0)
```
