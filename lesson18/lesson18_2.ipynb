```python
import pandas as pd

df = pd.read_excel('各鄉鎮市區人口密度.xlsx',index_col=0)
df
```

一個DataFrame,如何將DataFrame做一個統計資料

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    return s.mean()
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    return pd.Series([s.mean()],index=['平均'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    return pd.Series([s.mean(),s.max(),s.min()],index=['平均','最高','最低'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    return pd.Series([s.mean(),s.max(),s.min(),s.std()],index=['平均','最高','最低','標準差'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    print(s.max())
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    print(s == s.max())
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    print(s[s == s.max()])
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    print(s[s == s.max()].index)
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    print(s[s==s.max()].index[0])
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])
df1.apply(statistic,axis=0)
```

```python
df1 = df[["年底人口數","土地面積","人口密度"]]

def statistic(s):
    max_city = s[s == s.max()].index[0]
    min_city = s[s == s.min()].index[0]
    return pd.Series([s.mean(),s.max(),max_city,s.min(),min_city,s.std(),s.median()],
                     index=['平均','最高','最高縣市','最低','最低縣市','標準差','中間數'])

df1.apply(statistic,axis=0)
```
